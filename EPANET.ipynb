```python
import os
from  epanettools.epanettools import EPANetSimulation, Node
import matplotlib.pyplot as plt
```

```python
os.chdir("E://Presentations//2016_python//examples")
es=EPANetSimulation("Net3.inp")
es.run()
p=Node.value_type['EN_PRESSURE']
```

```python
%matplotlib inline
for n in es.network.nodes.values():
    plt.plot(es.network.nodes[n.id].results[p], label=n.id)
```

```python
%matplotlib inline
for n in list(es.network.nodes.values())[:10]:
    plt.plot(es.network.nodes[n.id].results[p], label=n.id)
plt.legend(loc='upper left')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Pressure Head (m)')
plt.xlabel('Timestep')
```
